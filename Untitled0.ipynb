```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('f6abd377-0bc2-4951-9a3d-216daa114cbd.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply Sobel edge detection
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = cv2.bitwise_or(sobelx, sobely)

# Apply Canny edge detection
canny = cv2.Canny(image, 100, 200)

# Plotting
plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2), plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection'), plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3), plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection'), plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.savefig('edge_detection_results.png')
plt.show()
```

```python
import cv2
import numpy as np

# Function to add salt and pepper noise
def add_salt_pepper_noise(image, salt_vs_pepper=0.5, amount=0.004):
    noisy_image = np.copy(image)
    num_salt = np.ceil(amount * image.size * salt_vs_pepper)
    num_pepper = np.ceil(amount * image.size * (1. - salt_vs_pepper))

    # Add salt noise
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy_image[tuple(coords)] = 255

    # Add pepper noise
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy_image[tuple(coords)] = 0

    return noisy_image

# Function to add gaussian noise
def add_gaussian_noise(image):
    row, col, ch = image.shape
    mean = 0
    var = 0.1
    sigma = var**0.5
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    gauss = (gauss - np.min(gauss)) / (np.max(gauss) - np.min(gauss)) * 255
    gauss = gauss.astype(np.uint8)  # Convert to same type as image
    noisy_image = cv2.add(image, gauss)
    return noisy_image


# Load the original image
image = cv2.imread('f6abd377-0bc2-4951-9a3d-216daa114cbd.jpeg')

# Ensure the image is in the correct format
if image is None:
    print('Image not found')
else:
    image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    # Add salt and pepper noise
    sp_noisy_image = add_salt_pepper_noise(image)

    # Add gaussian noise
    gaussian_noisy_image = add_gaussian_noise(image)

    # Save the noisy images
    cv2.imwrite('salt_pepper_noisy_image.jpg', sp_noisy_image)
    cv2.imwrite('gaussian_noisy_image.jpg', gaussian_noisy_image)

    print('Noise addition completed.')
```

```python
from PIL import Image
import numpy as np

# Load the original image
image_path = 'f6abd377-0bc2-4951-9a3d-216daa114cbd.jpeg'
image = Image.open(image_path)

# Rotate the image by 90 degrees
rotated_image = image.rotate(90)

# Flip the image vertically
flip_vertical_image = image.transpose(Image.FLIP_TOP_BOTTOM)

# Flip the image horizontally
flip_horizontal_image = image.transpose(Image.FLIP_LEFT_RIGHT)

# Save the transformed images
rotated_image.save('rotated_image.jpg')
flip_vertical_image.save('flip_vertical_image.jpg')
flip_horizontal_image.save('flip_horizontal_image.jpg')

print('Geometric transformations completed.')
```
